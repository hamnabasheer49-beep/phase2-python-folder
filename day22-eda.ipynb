## Step 1: Load and Inspect the Dataset

## Step 3: Data Exploration using GroupBy and Filtering

Here, I explore the dataset to understand patterns between gender, income, and spending behavior.

### Average Spending Score by Gender

Here, I calculate the average spending score for each gender to understand if there is any difference in spending behavior between male and female customers.

```python
df.groupby("Gender")["Spending Score (1-100)"].mean()
```

### Average Annual Income by Gender

This analysis helps to compare the average income of male and female customers.

```python
df.groupby("Gender")["Annual Income (k$)"].mean()
```

### High Spending Customers

Here, I filter customers who have a spending score greater than 80 to identify high-value customers.

```python
df[df["Spending Score (1-100)"] > 80]
```

### High Income Customers

This step identifies customers with annual income greater than 70k.

```python
df[df["Annual Income (k$)"] > 70]
```

## Step 4: Data Visualization

In this step, I create visualizations to better understand relationships and distributions in the data.

```python
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"])
plt.title("Income vs Spending Score")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.show()
```

The scatter plot shows how customer spending varies with income, revealing different spending patterns among customers.

```python
plt.hist(df["Age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows the distribution of customer ages, indicating that most customers belong to a specific age group.

Customers with higher income do not always spend more, showing varied spending behavior.
