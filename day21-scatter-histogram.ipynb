```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Aisha", "Rahul", "John", "Fatima", "Arjun", "Sara", "David", "Meera"],
    "Age": [20, 21, 19, 22, 20, 21, 23, 20],
    "Gender": ["F", "M", "M", "F", "M", "F", "M", "F"],
    "Marks": [85, 78, 90, 88, 76, 95, 89, 92],
    "Hours_Studied": [3, 2, 4, 3, 2, 5, 4, 5]
}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df["Hours_Studied"],df["Marks"])
plt.title("Hours_studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```

Observation:
The scatter plot shows a positive relationship between hours studied and marks.
As the number of hours studied increases, the marks also tend to increase.
This indicates that studying more hours may lead to better performance.

```python
plt.hist(df["Marks"])
plt.title("Marks distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

Observation:
The distribution of marks is slightly concentrated in the higher range.
Most students have marks between 85 and 95, while fewer students have lower marks.
The data is moderately spread out and there are no clear outliers.
