```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\acer\OneDrive\Desktop\day03 messy dataset.csv")
```

```python
print(df)
```

```python
## Checking missing values to understand which columns need cleaning
df.isnull().sum()
```

```python
## Dropping rows where 'product' is missing because it is a key identifier
df=df.dropna(subset=["Product"])
print(df)
```

```python
# Filling missing 'Customer_Name' with 'Unknown' as a default category
df["Customer_Name"]=df["Customer_Name"].fillna("Unknown")
df
```

```python
# Removing duplicate rows to ensure data consistency
df=df.drop_duplicates()
df
```

```python
#to check the data types
df.dtypes
```

```python
#To convert data type of 'Order_Date' to datetime for correct analysis
df["Order_Date"] = pd.to_datetime(df["Order_Date"],errors="coerce")
print(df)
```

```python
#To fill the missing values in 'Order_Date'
df["Order_Date"]=df["Order_Date"].fillna(method="ffill")
df.info()
```
