```python
scores=[45,67,89,34,76,50,92,38,60,81]
```

```python
for score in scores:
    if score>=50:
        print(score,"Pass")
    else:
        print(score,"Fail")
    
```

```python
pass_count=0
fail_count=0
for score in scores:
    if score>=50:
       pass_count+=1
    else:
       fail_count+=1
```

```python
print(pass_count,"students passed out of 10")
print(fail_count,"students failed")
```

```python
pass_mark=60
for score in scores:
    if score>=pass_mark:
        print(score,"pass")
        
```
