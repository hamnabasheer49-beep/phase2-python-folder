```python
import pandas as pd

data = {
    'Product': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Marker', 'Scale', 'Bag', 'Bottle'],
    'Category': ['Stationery', 'Stationery', 'Stationery', 'Stationery', 'Stationery', 'Stationery', 'Accessories', 'Accessories'],
    'Sales': [200, 100, 600, 50, 300, 150, 1200, 800],
    'Profit': [50, 20, 150, 10, 80, 30, 300, 200],
    'Region': ['South', 'North', 'South', 'East', 'West', 'North', 'South', 'East']
}

df = pd.DataFrame(data)
df
```

```python
df[['Product','Sales']]
#This shows only product names and their sales values
```

```python
df[df['Sales']>df['Sales'].mean()]
#This shows products with sales higher than the average sales value.
```

```python
df.sort_values(by='Sales',ascending=False)
#This sorts the dataset from highest to lowest sales.
```

```python
df[(df['Sales']>200) & (df['Profit']>50)]
#This shows products that have both high sales and high profit.
```
