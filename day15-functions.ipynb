```python
#celsius to fahrenheit
def celsius_to_fahrenheit(temp):
    return(temp*9/5)+32
    
```

```python
print(celsius_to_fahrenheit(0))
print(celsius_to_fahrenheit(25))
print(celsius_to_fahrenheit(100))
```

```python
#grade calculator
def get_grade(score):
    if score>=90:
        return "A"
    elif score>=75:
        return "B"
    elif score>=50:
        return "C"
    else: 
        return "F"
```

```python
print(get_grade(95))
print(get_grade(80))
print(get_grade(65))
print(get_grade(45))
print(get_grade(50))
```
